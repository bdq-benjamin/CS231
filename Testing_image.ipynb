# Model LogisticRegression

## Image 1

```python
dir_img1 = '/content/drive/MyDrive/Năm 2/CV/Do An/Dataset/test/happy/PrivateTest_10077120.jpg'
```

```python
model_lr = load('/content/drive/MyDrive/Năm 2/CV/Do An/Models/model_lr.h5')
```

```python
img1 = cv2.imread(dir_img1)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB));
```

```python
img = cv2.imread(dir_img1,0)
img = cv2.resize(img, (48, 48))
(hog, hog_image) = feature.hog(img, orientations=9,pixels_per_cell=(6, 6), cells_per_block=(3, 3),block_norm='L2-Hys', visualize=True, transform_sqrt=True)
plt.title("Hog image")
plt.imshow(hog_image, cmap='gray');
```

```python
hog.shape
img1_test = hog.reshape(1,-1)
```

```python
y_pred = model_lr.predict(img1_test)
```

```python
y_pred
```

```python
label_map = {
    0: 'angry',
    1: 'disgust',
    2: 'fear',
    3: 'happy',
    4: 'neutral',
    5: 'sad',
    6: 'surprise'
}
```

```python
print('Predict:',label_map[int(y_pred)])
```
